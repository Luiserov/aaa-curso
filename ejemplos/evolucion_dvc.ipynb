```python
import numpy as np
import matplotlib.pyplot as pl
pl.style.use('ggplot')


dvc = [10, 50, 100]
```

```python
pl.figure(figsize=(10, 5))
for d_vc in dvc:
    M = np.linspace(1, 10 * d_vc, 1_000)
    pl.loglog(M, np.float_power(M, d_vc) * np.exp(-M), label=f"d_vc = {d_vc}")
    
pl.legend()
pl.show()
```
